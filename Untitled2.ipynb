**In case of loss or lack of numerical data, it can be filled according to the average or median values of the original data, so that the original data will not be disrupted in mathematical operations.**

**Let's examine each category of competition iteratively. Let's fill in the lost height and weight values by calculating the height and weight environments of each category..**

```python
df_temp = df.copy()
height_weight = ['Height', 'Weight']
for i in pd.unique(df.Event):                 # The cycle continues within the competition categories.
  filtre_event = df_temp.Event == i           # Let's create a filter by competition category.
  filtre_value = df_temp[filtre_event]        # Let's filter the data by competition category.

  for j in height_weight:                                              # he cycle continues for all athletes.
    average = np.round(np.mean(filtre_value[j]), 2)
    if ~np.isnan(average):                                        # If there is an average in the competition category
      filtre_value[j] = filtre_value[j].fillna(average)
    else:                                                         # If there is no average in the competition category
      general_average = np.round(np.mean(df[j]), 2)               
      filtre_value[j] = filtre_value[j].fillna(general_average)

  df_temp[filtre_event] = filtre_value

df = df_temp.copy()
df.info()
```

```python
average_age = np.round(np.mean(df.Age), 2)
print('Average of Age: {}'.format(average_age))
df['Age'] = df['Age'].fillna(average_age)
df.info()
```

**pandas.get_dummies() is used for data manipulation. It converts categorical data into dummy or indicator variables.**
* **syntax:  pandas.get_dummies(data, prefix=None, prefix_sep=’_’, dummy_na=False, columns=None, sparse=False, drop_first=False, dtype=None)**

Parameters:

* data: whose data is to be manipulated.
* prefix: String to append DataFrame column names. Pass a list with length equal to the number of columns when calling get_dummies on a DataFrame. Default value is None.
* prefix_sep: Separator/delimiter to use if appending any prefix. Default is ‘_’
* dummy_na: It adds a column to indicate NaN values, default value is false, If false NaNs are ignored.
* columns: Column names in the DataFrame that needs to be encoded. Default value is None, If columns is None then all the columns with object or category dtype will be converted.
* sparse: It  specify whether the dummy-encoded columns should be backed by a SparseArray (True) or a regular NumPy array (False). default value is False.
* drop_first: Remove first level to get k-1 dummies out of k categorical levels.
* dtype: Data type for new columns. Only a single dtype is allowed. Default value is np.uint8.
Returns: Dataframe (Dummy-coded data)

# **Anomaly Detection**
* **An outlier is a data point that differs significantly from other observations.**
* **An outlier may be dependent on variables or indicate experimental error, or sometimes be excluded from the dataset.**
* **It should be noted that it can cause serious problems in statistical analysis.**

```python
import collections
def anomaly_detection (dataframe, feature):
    outlier_indices = []                      # I will find outliers and add them to the list.
    
    for i in feature:
        Q1 = np.percentile(dataframe[i], 25)      # The first 25%
        Q3 = np.percentile(dataframe[i], 75)      # Third 25%
        IQR = Q3 - Q1                             
        outlier_step = 1.5 * IQR
        # We will detect the outlier value and its index.
        # Q1 - outlier_step == >> Returns the negative outlier.
        # Q3 + outlier_step == >> Returns the outlier in the positive direction.
        outlier_list_col = dataframe[(dataframe[i] < Q1 - outlier_step) | (dataframe[i] > Q3 + outlier_step)].index
        outlier_indices.extend(outlier_list_col)

    # We will find unique outliers.
    outlier_indices = collections.Counter(outlier_indices)
    # If an example has a different value in the v number column, let's consider this as an inconsistency. (We will choose those with at least 2 outliers)
    multiple_outliers = list(i for i, v in outlier_indices.items() if v > 1)

    return multiple_outliers
```

```python
data_anomaly = df.loc[anomaly_detection(df, ['Age', 'Weight', 'Height'])]
data_anomaly.Sport.value_counts()
```

```python
plt.figure()
plt.bar(data_anomaly.Sport.value_counts().index, data_anomaly.Sport.value_counts().values)
plt.grid(True)
plt.xticks(rotation = 45)
plt.show()
```

# **Data Analysis in Time Series**
* Average Age, Height and Weight Change by Years
* Number of Medals by Years

```python
data_time = df.copy()
data_time.head(2)
```

```python
data_time.Year.unique()
```

```python
np.sort(data_time.Year.unique())
```

* We will convert the year values in the dataframe to datetime data type.

```python
pd.to_datetime(data_time['Year'], format = '%Y')
```

```python
data_time['Data_Time'] = pd.to_datetime(data_time['Year'], format = '%Y')
data_time.head(3)
```

```python
data_time = data_time.set_index('Data_Time')
data_time.drop(['Year'], axis = 1, inplace = True)
data_time
```

```python
sorted_data = data_time.resample('2A').mean()
sorted_data
```

```python
sorted_data.dropna(axis = 0, inplace = True)
sorted_data.head()
```

```python
plt.figure()
sorted_data.plot()
plt.title('Average Age, Height and Weight Change by Years')
plt.xlabel('Year')
plt.grid(True)
plt.show()
```

```python
data_time = pd.get_dummies(data_time, columns=['Medal'])
data_time.head(2)
```

```python
sorted_data = data_time.resample('2A').sum()
sorted_data = sorted_data[~(sorted_data == 0).any(axis=1)]
sorted_data.tail()
```

```python
plt.figure()
sorted_data.loc[:, ['Medal_Gold', 'Medal_Silver', 'Medal_Bronze']].plot()
plt.xlabel('Year')
plt.ylabel('Number')
plt.grid(True)
plt.show()
```
